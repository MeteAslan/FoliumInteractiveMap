```python
import pandas as pd
```

```python
import folium
from folium import Marker
from folium.plugins import MarkerCluster
```

```python
data = pd.read_csv("Toronto_apartment_rentals_2018.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Bathroom'] = data['Bathroom'].astype(int)
```

```python
data['Price'] = data['Price'].str.replace('[\$,]', '').astype(float)
```

```python
data.head()
```

```python
def map_creator(data):
  map = folium.Map(location=[43.6532, -79.3832],zoom_start=10)
  mc = MarkerCluster()
  for idx, row in data.iterrows():
    mc.add_child(Marker(
        [row['Lat'], row['Long']],
        popup="<p><b>Bedroom</b>: {}</p><p><b>Bathroom</b>: {}</p><p><b>Adress</b>: {}</p><p><b>Price</b>: ${}</p>".format(row['Bedroom'],row['Bathroom'],row['Address'],row['Price'])                            
    ))
  mc.add_to(map)  
  return map  
```

```python
print("En Yüksek Daire Fiyatı: ${}\nEn Düşük Daire Fiyatı: ${}".format(data['Price'].max(),data['Price'].min()))

budget = float(input("Bütçenizi Giriniz: "))

data_2 = data[data['Price'] <= budget]
```

```python
from geopy.geocoders import Nominatim
```

```python
def loc_finder(search):
  try:
    geolocator = Nominatim(user_agent="project")
    location = geolocator.geocode(search)
    return location
  except:
    print("Location Not Found")    
  
```

```python
search = input("Hangi Bölgede Ev Bakıyorsunuz: ")
loc = loc_finder(search)

map_1 = map_creator(data_2)

folium.CircleMarker(
    location=[loc.latitude, loc.longitude],
    radius=30,
    tooltip=loc.address,
    color="#3186cc",
    fill=True,
    fill_color="#3186cc",
).add_to(map_1)
```
